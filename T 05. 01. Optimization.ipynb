```python
import tensorflow as tf

stock_times = [0, 1, 2, 3, 4]
stock_prices = [1000, 1001, 1002, 1003, 1004]

count = len(stock_prices)

x = tf.placeholder("float",name='feature')
y = tf.placeholder("float",name='output')
w = tf.Variable([0.8], name='weight')
b = tf.Variable([2.0], name='bias')
model = tf.add(tf.multiply(w, x) , b, name='model')
#model = W * x + b

# Minimize squared errors
#cost = tf.reduce_sum(tf.square(model - y))/(2 * count) #L2 loss
squared_deltas = tf.square(model - y)
cost = tf.reduce_sum(squared_deltas)

optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(cost)

with tf.Session() as session:
    sess.run(tf.global_variables_initializer())
    # Train 1000 times.
    for _ in range(1000):
        sess.run(train, {x: stock_times, y: stock_prices})
        result = sess.run([w, b])
        print("i: {i:.3f} w: {w:.3f} b: {b:.3f}".format(i=i,w=result[0][0],b=result[1][0]))
    
    # Get optimized values for "W" and "b" with trained model.
    result = sess.run([w, b])
    print(result)
    
```
